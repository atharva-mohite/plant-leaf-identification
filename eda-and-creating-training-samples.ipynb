**Preprocessing the images in the datasets and appending the arrays with the corresponding labels**

```python
# Creating lists to store image arrays and their corresponding labels
data = []
labels = []

path="../input/plantleafidentification/Leaves"

for image in os.listdir(path):
    
    image_path = path+"/"+image
    image = image.replace(".jpg", "")
    
    l=0
    
    for j in range(1,32):
        if int(image) in label_dict[j]:
            l = j
    
    labels.append(l) # Appending the label to the list 'labels'
    
    img = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)
    img = cv2.resize(img, (Img_size, Img_size),
               interpolation = cv2.INTER_NEAREST)
    
    array = img_to_array(img) # Converting the resized image to array
    data.append(array) # Appending the array to the list 'data'
    
    
# Function to blur an image using open cv
    blurred_image = cv2.blur(src=img, ksize=(5, 5))
    array2 = img_to_array(blurred_image)
    data.append(array2)
    labels.append(l)
```
